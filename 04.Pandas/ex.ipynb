```python
d = {
    "age":np.random.randint(1, 100, size=30),
    "tall":np.round(np.random.normal(170, 10, size=30),2)
}
df = pd.DataFrame(d)
df
```

right: 구간경계의 오른쪽(True-기본)을 포함할지 왼쪽(False)을 포함할지

['소' < '중' < '대']

```python
a = pd.cut(df['age'],
          bins = 3,
          right = True,
          labels = ['하', '중', '상']
          )
a
```

```python
df['나이대'] =a
df
```

```python
df.groupby('나이대')['tall'].count()
```
